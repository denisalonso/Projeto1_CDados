```python
#    === FILTRO ===


#referencias para o filtro
stop_words = set(stopwords.words('english'))


#funcao cleanup
def cleanup(txt):
    new_txt = ''
    clean_txt = txt.replace('\\n',' ')
    split_txt = clean_txt.split(' ')
    for word in split_txt:
        if 'https' not in word:
            new_word = ''
            for character in word:
                if character.lower() in 'abcdefghijklmnopqrstuvwxyz0123456789':
                    new_word += character.lower()
            if new_word != '':
                new_txt += ' '
                new_txt += new_word
    word_tokens = word_tokenize(new_txt)
    texto = [w for w in word_tokens if not (w.lower() in stop_words)]
    
    return texto
```

```python
#aplicando o filtro

comentarios_fisica = dados_fisica.Comment.apply(cleanup)
comentarios_quimica = dados_quimica.Comment.apply(cleanup)
comentarios_biologia = dados_biologia.Comment.apply(cleanup)

comentarios_fisica
```

```python
#construindo os livros

livro_fisica = []
livro_biologia = []
livro_quimica = []
livrao = ''


for e in comentarios_fisica:
    for palavra in e:
        livro_fisica.append(palavra)
for e in comentarios_quimica:
    for palavra in e:
        livro_quimica.append(palavra)
for e in comentarios_biologia:
    for palavra in e:
        livro_biologia.append(palavra)
livrao = livro_fisica + livro_quimica + livro_biologia   

print(livro_fisica)
```

```python
#transformando os livros em series


fisica_dados = pd.Series(livro_fisica)
quimica_dados = pd.Series(livro_quimica)
biologia_dados = pd.Series(livro_biologia)
dados_total = pd.Series(livrao)

fisica_dados
```

```python
#pegando valores brutos e relativos das palavras


fisica_absoluto = fisica_dados.value_counts()
quimica_absoluto = quimica_dados.value_counts()
biologia_absoluto = biologia_dados.value_counts()
total_absoluto = dados_total.value_counts()

fisica_relativo = fisica_dados.value_counts(True)
quimica_relativo = quimica_dados.value_counts(True)
biologia_relativo = biologia_dados.value_counts(True)
total_relativo = dados_total.value_counts(True)

biologia_absoluto
print(len(biologia_dados))
```

```python
#probabilidade das materias

P_fisica = len(fisica_dados)/len(dados_total)
P_quimica = len(quimica_dados)/len(dados_total)
P_biologia = len(biologia_dados)/len(dados_total)

print(f'Porbabilidade fisica: {P_fisica}\nProbabilidade quimica: {P_quimica}\nProbabilidade biologia: {P_biologia}\ntotal: {P_fisica + P_quimica + P_biologia}')
```

```python
# === LA PLACE ===

# para melhorar a precisao do nosso classificador, aplicaremos o Laplace, para evitar multiplicações por 0,
# pois caso a palavra nao exista em nenhum dos livros, a probabilidade dela existir seria 0, o que prejudicaria nossos calculos




freq_fisica_total = fisica_absoluto.sum()
freq_quimica_total = quimica_absoluto.sum()
freq_biologia_total = biologia_absoluto.sum()
```

```python
# === FUNCAO CLASSIFICADORA ===

def classifica(comentario):
    P_fisica_frase = P_fisica
    P_quimica_frase = P_quimica
    P_biologia_frase = P_biologia
    
    for palavra in comentario:
        if palavra in fisica_absoluto:
            freq_palavra_fisica = fisica_absoluto[palavra]
            
            f = (freq_palavra_fisica + 1)/(freq_fisica_total + len(total_absoluto))
        else:
            f = 1/(len(fisica_dados) + freq_fisica_total)
        if palavra in quimica_absoluto:
            freq_palavra_quimica = quimica_absoluto[palavra]
            q = (freq_palavra_quimica + 1)/(freq_quimica_total + len(total_absoluto))
        else:
            q = 1/(len(quimica_dados) + freq_quimica_total)
        if palavra in biologia_absoluto:
            freq_palavra_biologia = biologia_absoluto[palavra]
            b = (freq_palavra_biologia + 1)/(freq_biologia_total + len(total_absoluto))
        else:
            b = 1/(len(biologia_dados) + freq_biologia_total)
            
        P_fisica_frase *= f
        P_quimica_frase *= q
        P_biologia_frase *= b
    if P_fisica_frase > P_quimica_frase and P_fisica_frase > P_biologia_frase:
        return 'Physics'
    elif P_quimica_frase > P_fisica_frase and P_quimica_frase > P_biologia_frase:
        return 'Chemistry'
    elif P_biologia_frase > P_quimica_frase and P_biologia_frase > P_quimica_frase:
        return 'Biology'
#     for palavra in comentario:
#         if palavra in fisica_absoluto:
#             P_fisica_frase *= total_relativo[palavra]
#         if palavra in quimica_absoluto:
#             P_quimica_frase *= total_relativo[palavra]
#         if palavra in biologia_absoluto:
#             P_biologia_frase *= total_relativo[palavra]
            
            
#         if palavra not in fisica_absoluto and palavra not in quimica_absoluto and palavra not in biologia_absoluto:
#             P_fisica_frase *= isoladas_relativo
#             P_quimica_frase *= isoladas_relativo
#             P_biologia_frase *= isoladas_relativo
# #         if palavra not in quimica_absoluto:
# #             P_quimica_frase *= isoladas_relativo
# #         if palavra not in biologia_absoluto:
# #             P_biologia_frase *= isoladas_relativo
#     if P_fisica_frase > P_quimica_frase and P_fisica_frase > P_biologia_frase:
#         return 'Physics'
#     elif P_quimica_frase > P_fisica_frase and P_quimica_frase > P_biologia_frase:
#         return 'Chemistry'
#     elif P_biologia_frase > P_quimica_frase and P_biologia_frase > P_quimica_frase:
#         return 'Biology'
```

```python
# === CLASSIFICANDO ===



rotulado = []

for frase in test.Comment:
    frase_filtrada = cleanup(frase)
    print(frase_filtrada)
    classificacao = classifica(frase_filtrada)
    rotulado.append(classificacao)
    
print(rotulado)
```

```python
# === FRAME COMPARACAO ===

test['classificadora'] = rotulado

f = 0
for i in range(len(test.Topic)):
    if test.Topic[i] == test.classificadora[i]:
        f += 1
print(f/len(test.Topic))   
test
```
